```python
import keras
```

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

import os
import random
```

```python
dog_folder = 'Animal_Images/dogs/'
cat_folder = 'Animal_Images/cats/'
```

```python
def sample_images(folder):
    img_paths = random.sample(os.listdir(folder), 5)
    imgs = []
    for path in img_paths:
        img = plt.imread(os.path.join(folder, path))
        imgs.append(img)
    fig, axis = plt.subplots(1,5)
    for index, image in enumerate(imgs):
        axis[index].imshow(image)
    
    plt.show()
```

```python
sample_images(dog_folder)
sample_images(cat_folder)
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator()
```

```python
train_gen = train_datagen.flow_from_directory(
    directory = "Animal_Images",
    target_size = (150,150),
    color_mode = 'grayscale',
    class_mode = 'binary',
    batch_size = 30000
)

print("Class indices:", train_gen.class_indices)
print("\n")
```

```python
from sklearn.model_selection import train_test_split
```

```python
data, targets = train_gen.next()
X_train, X_test, y_train, y_test = train_test_split(data, targets, test_size=0.2, random_state = 1)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_train.shape
```

```python
X_train[0]
```

```python
ndims = X_train.shape[1] * X_train.shape[2]
X_train = X_train.reshape(X_train.shape[0], ndims)
X_test = X_test.reshape(X_test.shape[0], ndims)

print("Training Shape:", X_train.shape)
print("Testing Shape:", X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
X_train[0]
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(100, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
y_train.shape
```

```python
model.fit(X_train, y_train, epochs=10, shuffle=True, verbose=2)
```

```python
model_loss, model_accuracy = model.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
from keras.layers import Conv2D,  Flatten
```

```python
model2 = Sequential()
model2.add(Conv2D(16, (3,3), activation='relu', input_shape=(32,22500,1)))
model2.add(Conv2D(32, (3, 3), activation='relu'))
model2.add(Conv2D(64, (3, 3), activation='relu'))
model2.add(Flatten())
model2.add(Dense(128, activation='relu'))
model2.add(Dense(1, activation='sigmoid'))
model2.summary()
```

```python
model2.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
X_train.shape
```
